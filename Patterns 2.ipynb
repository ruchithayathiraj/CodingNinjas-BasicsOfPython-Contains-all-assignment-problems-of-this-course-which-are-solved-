# Code : Inverted Number Pattern

Print the following pattern for the given N number of rows.
Pattern for N = 4
4444
333
22
1
Input format :
Integer N (Total no. of rows)
Output format :
Pattern in N lines
Constraints :
0 <= N <= 50
Sample Input 1:
5
Sample Output 1:
55555 
4444
333
22
1
Sample Input 2:
6
Sample Output 2:
666666
55555 
4444
333
22
1

```python
n=int(input())
i=1
while i<=n:
    j=1
    while j<= n-i+1:
        print(n-i+1,end='')
        j=j+1
    print()
    i=i+1
```

# Code : Mirror Number Pattern

Print the following pattern for the given N number of rows.
Pattern for N = 4


The dots represent spaces.


Input format :
Integer N (Total no. of rows)
Output format :
Pattern in N lines
Constraints
0 <= N <= 50

```python
n=int(input())
i=1
while i<=n:
    spaces=1
    while spaces<=n-i:
        print(' ',end='')
        spaces=spaces+1
    count=1
    while count<=i:
        print(count,end='')
        count=count+1
    print()
    i=i+1
```

# Code : Star Pattern

Print the following pattern
Pattern for N = 4




The dots represent spaces.



Input Format :
N (Total no. of rows)
Output Format :
Pattern in N lines
Constraints :
0 <= N <= 50
Sample Input 1 :
3
Sample Output 1 :
   *
  *** 
 *****
Sample Input 2 :
4
Sample Output 2 :
    *
   *** 
  *****
 *******

```python
n=int(input())
i=1
while i<=n:
    spaces=1
    while spaces<=n-i:
        print(' ',end='')
        spaces=spaces+1
    j=1
    while j<=i:
        print('*',end='')
        j=j+1
    p=i-1
    while p>=1:
        print('*',end='')
        p=p-1
    print()
    i=i+1
  
```

# Code : Triangle of Numbers

Pattern for N = 4

Input format :
Integer N (Total no. of rows)
Output format :
Pattern in N lines
Constraints :

# Zeros and Stars Pattern

Print the following pattern
Pattern for N = 4
*000*000*
0*00*00*0
00*0*0*00
000***000

Input Format :
N (Total no. of rows)
Output Format :
Pattern in N lines
Sample Input 1 :
3
Sample Output 1 :
*00*00*
0*0*0*0
00***00
Sample Input 2 :
5
Sample Output 2 :
*0000*0000*
0*000*000*0
00*00*00*00
000*0*0*000
0000***0000

```python
n = int(input())
rows = 2*n + 1
i = 1

while i <= n:
    j = 1
    while j <= rows:
        if(j == i or j == (rows // 2 + 1) or j == rows - i +1):
            print('*',end = '')
            
        else:
            print('0',end = '')

        j = j + 1
        
    print()
    i = i + 1
```

# Pyramid Number Pattern

Print the following pattern for the given number of rows.
Pattern for N = 4
   1
  212
 32123
4321234
Input format : N (Total no. of rows)

Output format : Pattern in N lines


```python
n = int(input())
i = 1

while i <= n:
    sp = 1
    while sp <= n-i:
        print(' ',end ='')
        sp = sp + 1
        
    p = i
    j = 1
    
    while j <= i:
        print(p, end ='')
        p = p - 1
        j = j + 1
        
    p = 2
    while p <= i:
        print(p, end ='')
        p = p + 1
    
    print()
    i = i + 1
```

# Arrow pattern

Print the following pattern for the given number of rows.
Assume N is always odd.
Note : There is space after every star.
Pattern for N = 7
*
 * *
   * * *
     * * * *
   * * *
 * *
*
Input format :
Integer N (Total no. of rows)
Output format :
Pattern in N lines
Sample Input :
11
Sample Output :
*
 * *
   * * *
     * * * *
       * * * * *
         * * * * * *
       * * * * *
     * * * *
   * * *
 * *
*

```python
n = int(input())
m = n // 2 + 1
i = 1

while i <= m:
    sp = 1
    while sp < i:
        print(" ", end = "")
        sp+=1
        
    j = i
    c = 1
    while j >= i and c <= i:
        print("* ",end = "")
        j+=2
        c+=1
        
    print()
    i+=1
   
i = m - 1   
while i >= 1: 
    sp = 1
    while sp < i:
            print(" ", end = "")
            sp+=1
            
    j = i
    c = 1
    while j >= 1 and c <= i:
        print("* ",end = "")
        j+=2
        c+=1
        
    print()
    i-=1
```
